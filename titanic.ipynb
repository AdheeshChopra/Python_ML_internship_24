```python
import numpy
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
%matplotlib inline
```

```python
auto=pd.read_csv("titanic.csv")
auto.head()
```

```python
sexvswho=auto.groupby(["sex","who"]).size().unstack(fill_value=0)
sexvswho
```

```python
sexvswho.plot(kind='bar',stacked=True)
plt.show()
```

```python
sexvswho.plot(kind='bar',stacked=False)
plt.show()
```

```python
data=pd.read_csv('titanic.csv')
plt.figure(figsize=(8, 6))
sns.histplot(data['age'].dropna(), bins=30, kde=True, color='red')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age ')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.set(rc={"axes.facecolor":"#FFFFFF","axes.grid":False,"xtick.labelsize":10,"ytick.labelsize":10})

plt.title("Sex VS Who")
plt.ylabel("Sex")
plt.xlabel("Who")
sns.lineplot(x=auto["sex"],y=auto["who"].index.values,color='r')

plt.show()
```
