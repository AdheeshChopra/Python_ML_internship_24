```python
import numpy
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
%matplotlib inline
```

```python
auto=pd.read_csv("Automobile_data.csv")
auto.head()
```

```python
plt.plot(auto["fuel-type"],auto["make"])
plt.show
```

```python
brandvsfuel=auto.groupby(["make","fuel-type"]).size().unstack(fill_value=0)
brandvsfuel
```

```python
brandvsfuel.plot(kind='bar',stacked=True)
plt.show()
```

```python
brandvsfuel.plot(kind='bar',stacked=True)
plt.show()
```

```python
brandvsdoor=auto.groupby(["make","num-of-doors"]).size().unstack(fill_value=0)
brandvsfuel
```

```python
brandvsdoor.plot(kind='bar',stacked=True)
plt.show()
```

```python
brandvsdoor.plot(kind='bar',stacked=False)
plt.show()
```

```python
brandvsdoorVShp=auto.groupby(["make","num-of-doors","horsepower"]).size().unstack(fill_value=0)
brandvsdoorVShp
```

```python
brandvsdoorVShp.plot(kind='bar',stacked=True)
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=auto["wheel-base"], linewidth=1.5, label="Wheel Base")
sns.lineplot(data=auto["engine-size"], linewidth=1.5, label="engine Size")

plt.title("Wheel Base VS Engine Size")
plt.ylabel("Wheel Base")
plt.xlabel("Engine Size")

plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.set(rc={"axes.facecolor":"#FFFFFF","axes.grid":False,"xtick.labelsize":10,"ytick.labelsize":10})

plt.title("Wheel Base VS Engine Size")
plt.ylabel("Wheel Base")
plt.xlabel("Engine Size")
plt.xticks(rotation=45)
sns.lineplot(x=auto["engine-size"],y=auto["wheel-base"].index.values,color='r')

plt.show()
```
